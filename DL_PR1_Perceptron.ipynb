```python
import numpy as np
def step_function(x):
  return 1 if x >= 0 else 0

def perceptron(x, w, b):
  y_in = np.dot(x, w) + b
  return step_function(y_in)

inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
weights = np.array([1, 1])
bias = -1.5

print("AND Gate")

for x in inputs:
  print(x, ":", perceptron(x, weights, bias))

weights = np.array([1, 1])
bias = -0.5

print("\nOR Gate")
for x in inputs:
  print(x, ":", perceptron(x, weights, bias))
inputs = np.array([[0], [1]])
weights = np.array([-1])
bias = 0.5

print("\nNOT Gate")
for x in inputs:
  print(x, ":", perceptron(x, weights, bias))
```
